```python
class Nodo:
  def __init__(self, valor):
    self.valor = valor
    self.derecha = None
    self.izquierda =  None


raiz = Nodo("*")

# izquierda
raiz.izquierda = Nodo("+")
raiz.izquierda.izquierda = Nodo(7)
raiz.izquierda.derecha = Nodo(3)

# derecha
raiz.derecha = Nodo("-")
raiz.derecha.derecha = Nodo(2)
raiz.derecha.izquierda = Nodo(5)


def recorrido_inorden(raiz):
    if raiz:
       recorrido_inorden(raiz.izquierda)
       print(raiz.valor, end=' ')
       recorrido_inorden(raiz.derecha)

print("\n RECORRIDO IN-ORDEN")
recorrido_inorden(raiz)
print("\n Resultado: ", (7 + 3)*(5- 2))


def recorrido_preorder(raiz):
   if raiz:
      print(raiz.valor, end=' ') 
      recorrido_preorder(raiz.izquierda)
      recorrido_preorder(raiz.derecha)

print("\n RECORRIDO PRE-ORDEN")
recorrido_preorder(raiz)
print("\n Resultado: ", 73 - 52)

def recorrido_postorden(raiz):
   if raiz:
        recorrido_postorden(raiz.izquierda)
        recorrido_postorden(raiz.derecha)
        print(raiz.valor, end= ' ')

print("\n RECORRIDO POST-ORDEN")
recorrido_postorden(raiz)
print("\n Resultado: ", 73 + 52)

```

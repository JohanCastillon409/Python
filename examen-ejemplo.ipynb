## Ejercicio 1: Gestión de Inventario (3 puntos)

Eres el encargado de una pequeña tienda de electrónicos. Tienes el siguiente inventario inicial:

```python
# Datos iniciales - NO MODIFICAR
productos = ["laptop", "mouse", "teclado", "monitor", "audífonos"]
cantidades = [5, 12, 8, 3, 15]
precios = [850.0, 25.5, 75.0, 320.0, 45.0]

print("=== INVENTARIO INICIAL ===")
for i in range(len(productos)):
    print(f"{productos[i]}: {cantidades[i]} unidades a ${precios[i]} c/u")
```

**Tu misión:**

a) Llegó un pedido de 20 cargadores a $18.0 cada uno. Agrégalo al inventario.

b) Un cliente compró 7 audífonos. Actualiza la cantidad.

c) El proveedor te ofreció un descuento: todos los productos que cuesten más de $50 tendrán un 10% de descuento.

d) Muestra **solo** los productos que tienen menos de 10 unidades en stock.

```python

productos = ["laptop", "mouse", "teclado", "monitor", "audífonos"]
cantidades = [5, 12, 8, 3, 15]
precios = [850.0, 25.5, 75.0, 320.0, 45.0]

print("=== INVENTARIO INICIAL ===")
for i in range(len(productos)):
    print(f"{productos[i]:<12}: {cantidades[i]} unidades a ${precios[i]:.2f} c/u")

# a) 
productos.append("cargadores")
cantidades.append(20)
precios.append(18.0)


# b) 

indice_audifonos = productos.index("audífonos")
cantidades[indice_audifonos] -= 7



# c) 
for i in range(len(precios)):
    if precios[i] > 50:
        precios[i] = precios[i] * 0.90

print("=== INVENTARIO FINAL ACTUALIZADO ===")
for i in range(len(productos)):
    print(f"{productos[i]:<12}: {cantidades[i]} unidades a ${precios[i]:.2f} c/u")
print("\n" + "="*50)

# d)
print("=== PRODUCTOS CON MENOS DE 10 UNIDADES ===")
print("="*45)
for i in range(len(productos)):
    if cantidades[i] < 10:
        print(f"{productos[i]:<12}: {cantidades[i]} unidades a ${precios[i]:.2f} c/u")
```

## Ejercicio 2: Sistema de Calificaciones (3 puntos)

Como profesor, necesitas procesar las calificaciones finales de tus estudiantes:

```python
# Datos iniciales - NO MODIFICAR
estudiantes = ["Ana", "Carlos", "Diana", "Eduardo", "Fernanda", "Gabriel"]
calificaciones = [85, 92, 78, 96, 83, 89]

print("=== CALIFICACIONES ACTUALES ===")
for i, estudiante in enumerate(estudiantes):
    print(f"{estudiante}: {calificaciones[i]}")
```

**Tu misión:**

a) Debido a un error en la corrección, todos los estudiantes recibirán 3 puntos adicionales.

b) Crea una lista con solo los nombres de los estudiantes que aprobaron (≥80).

c) Encuentra al estudiante con la calificación más alta y más baja (muestra nombre y calificación).

d) Calcula el promedio del grupo y muestra cuántos estudiantes están por encima del promedio.

```python
# Escribe tu solución aquí
estudiantes = ["Ana", "Carlos", "Diana", "Eduardo", "Fernanda", "Gabriel"]
calificaciones = [85, 92, 78, 96, 83, 89]

print("=== CALIFICACIONES INICIALES ===")
for i, estudiante in enumerate(estudiantes):
    print(f"{estudiante}: {calificaciones[i]}")


print("\n" + "="*30)
print("     SOLUCIÓN DEL EJERCICIO")
print("="*30 + "\n")

# a) 
calificaciones_actualizadas = [nota + 3 for nota in calificaciones]
print("a) Calificaciones con 3 puntos adicionales:")
for i, estudiante in enumerate(estudiantes):
    print(f"  - {estudiante}: {calificaciones_actualizadas[i]}")

# b) 
aprobados = [estudiantes[i] for i, nota in enumerate(calificaciones_actualizadas) if nota >= 80]
print("\nb) Estudiantes que aprobaron (calificación >= 80):")
print(f"  - {aprobados}")

# c) 
nota_mas_alta = max(calificaciones_actualizadas)
estudiante_nota_alta = estudiantes[calificaciones_actualizadas.index(nota_mas_alta)]

nota_mas_baja = min(calificaciones_actualizadas)
estudiante_nota_baja = estudiantes[calificaciones_actualizadas.index(nota_mas_baja)]

print("\nc) Calificación más alta y más baja:")
print(f"  - Calificación más alta: {estudiante_nota_alta} con {nota_mas_alta} puntos.")
print(f"  - Calificación más baja: {estudiante_nota_baja} con {nota_mas_baja} puntos.")

# d) 
promedio = sum(calificaciones_actualizadas) / len(calificaciones_actualizadas)
estudiantes_sobre_promedio = sum(1 for nota in calificaciones_actualizadas if nota > promedio)

print("\nd) Análisis del promedio del grupo:")
print(f"  - El promedio del grupo es: {promedio:.2f}")
print(f"  - Hay {estudiantes_sobre_promedio} estudiantes con una calificación por encima del promedio.")
    
    


```

## Ejercicio 3: Análisis de Ventas Trimestrales (4 puntos)

Trabajas en el área de análisis de una empresa que vende en 4 ciudades diferentes. Tienes los datos de ventas de los últimos 3 meses:

```python
# Datos iniciales - NO MODIFICAR
ciudades = ["México", "Guadalajara", "Monterrey", "Puebla"]
meses = ["Enero", "Febrero", "Marzo"]

# Matriz de ventas [ciudad][mes]
ventas = [
    [150000, 180000, 220000],  # México
    [120000, 140000, 160000],  # Guadalajara  
    [200000, 190000, 250000],  # Monterrey
    [80000, 95000, 110000]     # Puebla
]

print("=== VENTAS TRIMESTRALES ===")
print("Ciudad\t\tEnero\t\tFebrero\t\tMarzo")
for i in range(len(ciudades)):
    print(f"{ciudades[i]}\t\t${ventas[i][0]:,}\t\t${ventas[i][1]:,}\t\t${ventas[i][2]:,}")
```

**Tu misión:**

a) Calcula las ventas totales por ciudad (suma de los 3 meses) y muestra el ranking de ciudades por ventas.

b) Identifica qué mes tuvo las mejores ventas a nivel nacional (suma de todas las ciudades).

c) Encuentra la ciudad y mes específico donde se registró la venta más alta.

d) Calcula el crecimiento porcentual de cada ciudad entre enero y marzo.

```python
# Escribe tu solución aquí

ciudades = ["México", "Guadalajara", "Monterrey", "Puebla"]
meses = ["Enero", "Febrero", "Marzo"]

ventas = [
    [150000, 180000, 220000],  # México
    [120000, 140000, 160000],  # Guadalajara
    [200000, 190000, 250000],  # Monterrey
    [80000, 95000, 110000]     # Puebla
]

print("=== VENTAS TRIMESTRALES ===")
print(f"{'Ciudad':<12}\tEnero\t\tFebrero\t\tMarzo")
for i in range(len(ciudades)):
    print(f"{ciudades[i]:<12}\t${ventas[i][0]:,}\t\t${ventas[i][1]:,}\t\t${ventas[i][2]:,}")


print("\n" + "="*40)
print("     ANÁLISIS DE VENTAS TRIMESTRALES")
print("="*40 + "\n")


# a) 
ventas_por_ciudad = {ciudades[i]: sum(ventas[i]) for i in range(len(ciudades))}
ranking_ciudades = sorted(ventas_por_ciudad.items(), key=lambda item: item[1], reverse=True)

print("a) Ranking de ciudades por ventas totales:")
for i, (ciudad, total) in enumerate(ranking_ciudades):
    print(f"  {i+1}. {ciudad:<12} - ${total:,}")


# b)
ventas_por_mes = [sum(venta_mes) for venta_mes in zip(*ventas)]
venta_max_mes = max(ventas_por_mes)
mejor_mes = meses[ventas_por_mes.index(venta_max_mes)]

print("\nb) Mejor mes en ventas a nivel nacional:")
print(f"  - El mes con mayores ventas fue {mejor_mes} con un total de ${venta_max_mes:,}")


# c) 
venta_mas_alta = 0
ciudad_venta_alta = ""
mes_venta_alta = ""

for i, ciudad_ventas in enumerate(ventas):
    for j, venta_mes in enumerate(ciudad_ventas):
        if venta_mes > venta_mas_alta:
            venta_mas_alta = venta_mes
            ciudad_venta_alta = ciudades[i]
            mes_venta_alta = meses[j]

print("\nc) Registro de venta más alta:")
print(f"  - La venta más alta fue de ${venta_mas_alta:,} en {ciudad_venta_alta} durante el mes de {mes_venta_alta}.")


# d)
print("\nd) Crecimiento porcentual (Enero vs Marzo):")
for i, ciudad in enumerate(ciudades):
    ventas_enero = ventas[i][0]
    ventas_marzo = ventas[i][2]
    crecimiento = ((ventas_marzo - ventas_enero) / ventas_enero) * 100
    print(f"  - {ciudad:<12}: {crecimiento:.2f}%")

```

## Ejercicio 4: Analizador de Tres en Raya (4 puntos)

Has desarrollado un juego de Tres en Raya (Tic-Tac-Toe) y necesitas crear una función que determine el ganador. El tablero usa 'X', 'O', o ' ' (espacio) para casillas vacías:

```python
# Datos iniciales - NO MODIFICAR
# Tableros de prueba
tablero1 = [
    ['X', 'O', 'X'],
    ['O', 'X', 'O'],
    ['O', 'X', 'X']
]

tablero2 = [
    ['O', 'O', 'O'],
    ['X', 'X', ' '],
    [' ', ' ', 'X']
]

tablero3 = [
    ['X', 'O', 'X'],
    ['O', 'O', 'X'],
    ['O', 'X', 'O']
]

# Función para mostrar tablero
def mostrar_tablero(tablero):
    print("   0   1   2")
    for i in range(3):
        print(f"{i}  {tablero[i][0]} | {tablero[i][1]} | {tablero[i][2]}")
        if i < 2:
            print("  ---|---|---")
    print()

print("=== TABLEROS DE PRUEBA ===")
print("Tablero 1:")
mostrar_tablero(tablero1)
print("Tablero 2:")
mostrar_tablero(tablero2)
print("Tablero 3:")
mostrar_tablero(tablero3)
```

**Tu misión:**

a) Crea una función `verificar_ganador(tablero)` que retorne:
   - 'X' si gana X
   - 'O' si gana O  
   - 'Empate' si no hay casillas vacías y no hay ganador
   - 'Juego continúa' si hay casillas vacías y no hay ganador

b) La función debe verificar:
   - Filas (3 iguales horizontalmente)
   - Columnas (3 iguales verticalmente)
   - Diagonales (principal e inversa)

c) Prueba tu función con los 3 tableros proporcionados y muestra los resultados.

```python
# Escribe tu solución aquí

tablero1 = [
    ['X', 'O', 'X'],
    ['O', 'X', 'O'],
    ['O', 'X', 'X']
]

tablero2 = [
    ['O', 'O', 'O'],
    ['X', 'X', ' '],
    [' ', ' ', 'X']
]

tablero3 = [
    ['X', 'O', 'X'],
    ['O', 'O', 'X'],
    ['O', 'X', 'O']
]


def mostrar_tablero(tablero):
    print("  0   1   2")
    for i in range(3):
        print(f"{i} {tablero[i][0]} | {tablero[i][1]} | {tablero[i][2]}")
        if i < 2:
            print(" ---|---|---")
    print()

def verificar_ganador(tablero):
    for i in range(3):
        if tablero[i][0] == tablero[i][1] == tablero[i][2] and tablero[i][0] != ' ':
            return tablero[i][0]
        if tablero[0][i] == tablero[1][i] == tablero[2][i] and tablero[0][i] != ' ':
            return tablero[0][i]

    if tablero[0][0] == tablero[1][1] == tablero[2][2] and tablero[0][0] != ' ':
        return tablero[0][0]
    
    if tablero[0][2] == tablero[1][1] == tablero[2][0] and tablero[0][2] != ' ':
        return tablero[0][2]

    if any(' ' in fila for fila in tablero):
        return 'Juego continúa'
    else:
        return 'Empate'


print("=== ANÁLISIS DE TABLEROS DE TRES EN RAYA ===\n")

tableros_de_prueba = [tablero1, tablero2, tablero3]

for i, tablero_actual in enumerate(tableros_de_prueba):
    print(f"--- Tablero {i+1} ---")
    mostrar_tablero(tablero_actual)
    resultado = verificar_ganador(tablero_actual)
    print(f"Resultado: {resultado}\n")

```

## Ejercicio 5: Estación Meteorológica (3 puntos)

Administras una estación meteorológica que registra temperaturas cada 6 horas durante una semana:

```python
# Datos iniciales - NO MODIFICAR
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
horas = ["00:00", "06:00", "12:00", "18:00"]

# Matriz de temperaturas [día][hora]
temperaturas = [
    [12, 18, 25, 22],  # Lunes
    [14, 20, 28, 24],  # Martes
    [10, 16, 23, 19],  # Miércoles
    [13, 19, 27, 25],  # Jueves
    [15, 21, 30, 26],  # Viernes
    [16, 22, 32, 28],  # Sábado
    [11, 17, 24, 20]   # Domingo
]

print("=== TEMPERATURAS DE LA SEMANA (°C) ===")
print("Día\t\t00:00\t06:00\t12:00\t18:00")
for i in range(len(dias)):
    print(f"{dias[i]}\t\t{temperaturas[i][0]}°\t{temperaturas[i][1]}°\t{temperaturas[i][2]}°\t{temperaturas[i][3]}°")
```

**Tu misión:**

a) Encuentra el día más caluroso y más frío de la semana (considera la temperatura máxima de cada día).

b) Determina a qué hora del día se registran generalmente las temperaturas más altas y más bajas.

c) Calcula la temperatura promedio de cada día y identifica qué días estuvieron por encima del promedio semanal.

d) Crea una lista con los días que tuvieron variaciones de temperatura superiores a 15°C (diferencia entre máxima y mínima del día).

```python
# Escribe tu solución aquí

dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
horas = ["00:00", "06:00", "12:00", "18:00"]

temperaturas = [
    [12, 18, 25, 22],  # Lunes
    [14, 20, 28, 24],  # Martes
    [10, 16, 23, 19],  # Miércoles
    [13, 19, 27, 25],  # Jueves
    [15, 21, 30, 26],  # Viernes
    [16, 22, 32, 28],  # Sábado
    [11, 17, 24, 20]   # Domingo
]

print("=== TEMPERATURAS DE LA SEMANA (°C) ===")
print(f"Día\t\t00:00\t06:00\t12:00\t18:00")
for i in range(len(dias)):
    print(f"{dias[i]:<10}\t{temperaturas[i][0]}°\t{temperaturas[i][1]}°\t{temperaturas[i][2]}°\t{temperaturas[i][3]}°")


print("\n" + "="*40)
print("     ANÁLISIS METEOROLÓGICO")
print("="*40 + "\n")

# a) 
maximas_diarias = [max(dia_temps) for dia_temps in temperaturas]

temp_mas_alta = max(maximas_diarias)
dia_mas_caluroso = dias[maximas_diarias.index(temp_mas_alta)]

temp_mas_baja_de_las_maximas = min(maximas_diarias)
dia_mas_frio = dias[maximas_diarias.index(temp_mas_baja_de_las_maximas)]

print("a) Día más caluroso y más frío (basado en la máxima diaria):")
print(f"  - Día más caluroso: {dia_mas_caluroso} ({temp_mas_alta}°C)")
print(f"  - Día más frío: {dia_mas_frio} ({temp_mas_baja_de_las_maximas}°C)")

# b) 
temps_por_hora = list(zip(*temperaturas))
promedios_por_hora = [sum(h_temps)/len(h_temps) for h_temps in temps_por_hora]

promedio_max_hora = max(promedios_por_hora)
hora_mas_calurosa = horas[promedios_por_hora.index(promedio_max_hora)]

promedio_min_hora = min(promedios_por_hora)
hora_mas_fria = horas[promedios_por_hora.index(promedio_min_hora)]

print("\nb) Horas con temperaturas generalmente más altas y bajas:")
print(f"  - Hora más calurosa (por promedio): {hora_mas_calurosa}")
print(f"  - Hora más fría (por promedio): {hora_mas_fria}")

# c) 
promedios_diarios = [sum(dia_temps)/len(dia_temps) for dia_temps in temperaturas]
total_temps = [temp for dia_temps in temperaturas for temp in dia_temps]
promedio_semanal = sum(total_temps) / len(total_temps)
dias_encima_promedio = [dias[i] for i, prom in enumerate(promedios_diarios) if prom > promedio_semanal]

print("\nc) Promedios de temperatura:")
print(f"  - Promedio semanal: {promedio_semanal:.2f}°C")
print("  - Días por encima del promedio semanal:")
print(f"    {dias_encima_promedio}")

# d) 
dias_alta_variacion = []
for i, dia_temps in enumerate(temperaturas):
    variacion = max(dia_temps) - min(dia_temps)
    if variacion > 15:
        dias_alta_variacion.append(dias[i])

print("\nd) Días con variación de temperatura superior a 15°C:")
print(f"  - {dias_alta_variacion}")
```

## Ejercicio 6: Desafío Integrador (3 puntos)

Para este último ejercicio, vas a combinar operaciones con arreglos 1D y 2D. Tienes datos de una competencia deportiva:

```python
# Datos iniciales - NO MODIFICAR
equipos = ["Águilas", "Tigres", "Leones", "Pumas"]
deportes = ["Futbol", "Basquet", "Voley"]

# Puntos obtenidos por equipo en cada deporte [equipo][deporte]
puntos_deportes = [
    [85, 72, 90],  # Águilas
    [78, 88, 65],  # Tigres  
    [92, 79, 88],  # Leones
    [88, 91, 77]   # Pumas
]

# Puntos extra por fair play (una dimensión)
fair_play = [15, 8, 12, 20]

print("=== PUNTOS POR DEPORTE ===")
print("Equipo\t\tFutbol\tBasquet\tVoley")
for i in range(len(equipos)):
    print(f"{equipos[i]}\t\t{puntos_deportes[i][0]}\t{puntos_deportes[i][1]}\t{puntos_deportes[i][2]}")
    
print("\n=== PUNTOS FAIR PLAY ===")
for i in range(len(equipos)):
    print(f"{equipos[i]}: {fair_play[i]} puntos")
```

**Tu misión:**

a) Calcula el puntaje total de cada equipo (suma de deportes + fair play) y determina el ganador.

b) Encuentra qué deporte tuvo los puntajes más altos en promedio y cuál el más bajo.

c) Crea una tabla de clasificación final ordenada de mayor a menor puntaje total.

d) Identifica si algún equipo fue el mejor en al menos 2 deportes individuales (sin contar fair play).

```python
# Escribe tu solución aquí

equipos = ["Águilas", "Tigres", "Leones", "Pumas"]
deportes = ["Fútbol", "Básquet", "Vóley"]

puntos_deportes = [
    [85, 72, 90],   # Águilas
    [78, 88, 65],   # Tigres
    [92, 79, 88],   # Leones
    [88, 91, 77]    # Pumas
]

fair_play = [15, 8, 12, 20]

print("=== PUNTOS POR DEPORTE ===")
print(f"{'Equipo':<10}\tFútbol\tBásquet\tVóley")
for i in range(len(equipos)):
    print(f"{equipos[i]:<10}\t{puntos_deportes[i][0]}\t{puntos_deportes[i][1]}\t{puntos_deportes[i][2]}")

print("\n=== PUNTOS FAIR PLAY ===")
for i in range(len(equipos)):
    print(f"{equipos[i]:<10}: {fair_play[i]} puntos")


print("\n" + "="*40)
print("     ANÁLISIS DE COMPETENCIA DEPORTIVA")
print("="*40 + "\n")

# a) 
puntajes_totales = [sum(puntos_deportes[i]) + fair_play[i] for i in range(len(equipos))]
puntaje_maximo = max(puntajes_totales)
ganador = equipos[puntajes_totales.index(puntaje_maximo)]

print("a) Ganador de la competencia:")
print(f"  - El equipo ganador es '{ganador}' con {puntaje_maximo} puntos totales.")

# b)
puntos_por_deporte = list(zip(*puntos_deportes))
promedios_por_deporte = [sum(deporte_puntos) / len(deporte_puntos) for deporte_puntos in puntos_por_deporte]

promedio_mas_alto = max(promedios_por_deporte)
deporte_promedio_alto = deportes[promedios_por_deporte.index(promedio_mas_alto)]

promedio_mas_bajo = min(promedios_por_deporte)
deporte_promedio_bajo = deportes[promedios_por_deporte.index(promedio_mas_bajo)]

print("\nb) Análisis de deportes por puntaje promedio:")
print(f"  - Deporte con mayor puntaje promedio: {deporte_promedio_alto} ({promedio_mas_alto:.2f})")
print(f"  - Deporte con menor puntaje promedio: {deporte_promedio_bajo} ({promedio_mas_bajo:.2f})")

# c) 
tabla_clasificacion = list(zip(equipos, puntajes_totales))
tabla_clasificacion.sort(key=lambda item: item[1], reverse=True)

print("\nc) Tabla de clasificación final:")
print(f"  {'Pos.':<5} {'Equipo':<10} {'Puntaje Total':<15}")
for i, (equipo, puntaje) in enumerate(tabla_clasificacion):
    print(f"  {i+1:<5} {equipo:<10} {puntaje:<15}")

# d) 
victorias_individuales = [0] * len(equipos)
for i, deporte_puntos in enumerate(puntos_por_deporte):
    max_puntos_deporte = max(deporte_puntos)
    ganador_deporte_idx = deporte_puntos.index(max_puntos_deporte)
    victorias_individuales[ganador_deporte_idx] += 1

equipos_dominantes = [equipos[i] for i, victorias in enumerate(victorias_individuales) if victorias >= 2]

print("\nd) Equipos dominantes en deportes individuales:")
if equipos_dominantes:
    for equipo in equipos_dominantes:
        print(f"  - {equipo} fue el mejor en 2 o más deportes.")
else:
    print("  - Ningún equipo fue el mejor en 2 o más deportes.")
```

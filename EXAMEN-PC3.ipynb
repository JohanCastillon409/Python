```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.derecha = None
        self.izquierda = None


raiz = Nodo("+")
raiz.izquierda = Nodo("-")
raiz.izquierda.izquierda = Nodo("*")
raiz.izquierda.izquierda.izquierda = Nodo("+")
nodo_10 = raiz.izquierda.izquierda.izquierda.izquierda = Nodo(10)
nodo_20 =raiz.izquierda.derecha =    Nodo(20)
nodo_3 = raiz.izquierda.izquierda.derecha =    Nodo(3)
nodo_1 = raiz.izquierda.izquierda.izquierda.derecha = Nodo(1)
raiz.derecha = Nodo("*")
nodo_3_derecha =raiz.derecha.izquierda = Nodo(3)
nodo_2 = raiz.derecha.derecha = Nodo(2)


def preorden(raiz):
    """Recorrido preorden recursivo"""
    if raiz is None:
        return []
    return [str(raiz.valor)] + preorden(raiz.izquierda) + preorden(raiz.derecha)

def inorden(raiz):
    """Recorrido inorden recursivo"""
    if raiz is None:
        return []
    return inorden(raiz.izquierda) + [str(raiz.valor)] + inorden(raiz.derecha)

def postorden(raiz):
    """Recorrido postorden recursivo"""
    if raiz is None:
        return []
    return postorden(raiz.izquierda) + postorden(raiz.derecha) + [str(raiz.valor)]


def evaluar_arbol(raiz):
    """Evalúa el árbol recursivamente"""
    if raiz.izquierda is None and raiz.derecha is None:
        return int(raiz.valor)
    
    izquierdo = evaluar_arbol(raiz.izquierda)
    derecho = evaluar_arbol(raiz.derecha)
    
    if raiz.valor == '+':
        return izquierdo + derecho
    elif raiz.valor == '-':
        return izquierdo - derecho
    elif raiz.valor == '*':
        return izquierdo * derecho
    


print("====== CONSTRUCCION DEL ARBOL =======")
print(f"\n Arbol construido:")

resultado = evaluar_arbol(raiz)
print("===== EVALUACION =====")
print("Expresion: ((10 + 1)* 3 - 20) + (3 * 2)")
print("Rsultado: ", resultado)


print("\n ===== RECORRIDOS ======")
print(f"PreOrden: ")
print(preorden(raiz))
print(f"InOrden: ")
print(inorden(raiz))
print(f"PostOrden: ")
print(postorden(raiz))

print("\n EXPLICACION PASO A PASO")
print(f"({nodo_10.valor} + {nodo_1.valor} ) = {nodo_10.valor + nodo_1.valor}")
res = (nodo_10.valor + nodo_1.valor) * nodo_3.valor
print(f"{nodo_10.valor + nodo_1.valor} * {nodo_3.valor} = {res} ")
print(f"{res} - {nodo_20.valor} = {res - nodo_20.valor}")
print(f"{nodo_3_derecha.valor} * {nodo_2.valor} = {nodo_3_derecha.valor * nodo_2.valor} ")
res2 = nodo_3_derecha.valor * nodo_2.valor
print(f"{res - nodo_20.valor} +  {res2}  = {res - nodo_20.valor + res2} ")

```

```python
# kata.run

class Nodo:

    def __init__(self, valor):
        self.valor = valor
        self.izquierdo = None
        self.derecho = None


raiz = Nodo("A")
raiz.izquierdo = Nodo("T")
raiz.derecho = Nodo("R")
raiz.izquierdo.izquierdo = Nodo("A")
raiz.derecho.izquierdo = Nodo(".")
raiz.derecho.derecho = Nodo("U")
raiz.izquierdo.izquierdo.izquierdo = Nodo("K")
raiz.derecho.derecho.derecho = Nodo("N")




# recorrer el nodo
def recorrer(nodo):
    if nodo:
        recorrer(nodo.izquierdo)
        print(nodo.valor, end='')
        recorrer(nodo.derecho)

recorrer(raiz) 





# RECORRER EL ARBOL (PRE-ORDEN)

# def recorrerp(nodo):
#     if nodo:
#         print(nodo.valor, end='')
#         recorrerp(nodo.izquierdo)
#         recorrerp(nodo.derecho)

# recorrerp(raiz) 


# raiz3 = Nodo("K")
# raiz3.izquierdo = Nodo("A")
# raiz3.derecho = Nodo(".")
# raiz3.derecho.izquierdo = Nodo("R")

# raiz3.izquierdo.izquierdo = Nodo("T")
# raiz3.izquierdo.derecho = Nodo("A")


```

```python
# NUMEROS IMAGEN 

class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierdo = None
        self.derecho = None

raiz = Nodo(10)
raiz.izquierdo =    Nodo(13)
raiz.derecho = Nodo(100)
raiz.izquierdo.izquierda = Nodo(25)
raiz.izquierdo.derecho = Nodo(11)

# recorrer el nodo
def recorrer(nodo):
    if nodo:
        recorrer(nodo.izquierdo)
        print(nodo.valor)
        recorrer(nodo.derecho)

recorrer(raiz)
```

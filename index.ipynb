### UN ABB ES UN ARBOL GENEALOGICO pero con numeros
1. Cada nodo tiene como maximo 2 hijos
2. Izquierda: numeros menores
3. Derecha: numeros mayores

```python
"""
          10
         /  \
        5    15
       / \   / \
      3  7 12  20

"""
```

```python
# LECCION 1: creando un Nodo
# Un nodo es: como una caja
# guarda: el valor (numero)
# guarda: dos flechas apuntando a sus hijos


class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izq = None # numeros menores
        self.der = None # numeros mayores

raiz = Nodo(10)
print(f"Nodo creado con valor: {raiz.valor}")

# IZQUIERDA
raiz.izq = Nodo(5)
raiz.izq.der = Nodo(7)
raiz.izq.izq = Nodo(3)

# DERECHA
raiz.der = Nodo(15)
raiz.der.der = Nodo(20)
raiz.izq = Nodo(12)

# CREA UN NODO CON EL VALOR 50 Y MUESTRA SU VALOR
raiz = Nodo(50)

print(f"Nuevo nodo ingresado con valor {raiz.valor}")

```

```python
# CONECTANDO NODOS
# crear arbol pequeÃ±o
"""
     10
    /  \
   5   15
"""



class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierda = None 
        self.derecha = None 

raiz = Nodo(10)
raiz.derecha = Nodo(15) # 5 es menor de 10
raiz.izquierda = Nodo(5) # 15 es mayor que 10

print(f"Raiz: {raiz.valor}")
print(f"Hijo izquierdo: {raiz.izquierda.valor}")
print(f"Hijo derecho: {raiz.derecha.valor}")


```

```python
##Desafio 3:

# ðŸ—£ï¸ III . insertar valores

#regla:
# si el valor es menor -> ir a la izquierda
# si el valor es mayor -> ir a la derecha
# repetir hasta encontrar un espacio vacio


def insertar(raiz,valor):
    if raiz is None:
        return  Nodo(valor)
    # si el valor es menor
    if valor < raiz.valor:
        raiz.izquierda = insertar(raiz.izquierda, valor)
    else:
        raiz.derecha = insertar(raiz.derecha,valor)

    return raiz


"""
   10
   / \
  5  15
 /
 3
"""

raiz = None
raiz = insertar(raiz, 10)
raiz = insertar(raiz, 5)
raiz = insertar(raiz, 15)
raiz = insertar(raiz, 3)

print(f"Raiz: {raiz.valor}")
print(f"izquierda: {raiz.izquierda.valor}")
print(f"izquierda de izquierda: {raiz.izquierda.izquierda.valor}")
print(f"Derecha: {raiz.derecha.valor}")

```
